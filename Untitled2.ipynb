```python
predict_df = new_df[['incident_priority','incident_symptom_code','incident_resolution_time']]        
predict_df.head()
```

```python
continent_dummies = pd.get_dummies(predict_df.incident_priority, prefix='priority').iloc[:, 1:]
```

```python
symptom = pd.get_dummies(predict_df.incident_symptom_code, prefix='symp').iloc[:, 1:]
```

```python
final = pd.concat([predict_df, continent_dummies, symptom], axis=1)
```

```python
final.head()
```

```python
final.drop(columns=['incident_priority', 'incident_symptom_code'], inplace =True)
```

```python
final.head()
```

```python
y = final.iloc[:,0]
```

```python
X = final.iloc[:,1:]
```

```python
y=y.reshape(-1, 1)
y.shape
```

```python
y.shape, X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=1/3,  random_state=1)
# alpha=0 is equivalent to linear regression.
from sklearn.linear_model import Ridge

# Instantiate the model.
#(Alpha of zero has no regularization strength, essentially a basic linear regression.)
ridgereg = Ridge(alpha=0, normalize=True)

# Fit the model.
ridgereg.fit(X_train, y_train)

# Predict with fitted model.
y_pred = ridgereg.predict(X_test)
#print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
y_pred
```

```python
#BUILDING THE OPTIMAL MODEL USING BackwardElimination
import statsmodels.formula.api as sm
#X = np.append(arr = np.ones((50,1))),values =X ,axis=1)
X_opt = X[:,[0,1,2,3,4,5]]
regressor_OLS = sm.OLS(endog=y, exog =X_opt).fit()
regressor_OLS.summary() 
```
